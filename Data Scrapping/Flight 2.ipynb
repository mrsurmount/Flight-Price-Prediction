```python
import selenium
from selenium import webdriver
import time
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import NoSuchElementException
```

```python
driver = webdriver.Chrome('chromedriver.exe')
```

```python
url= "https://flight.yatra.com/air-search-ui/dom2/trigger?type=O&viewName=normal&flexi=0&noOfSegments=1&origin=DEL&originCountry=IN&destination=BOM&destinationCountry=IN&flight_depart_date=27%2F04%2F2022&ADT=1&CHD=0&INF=0&class=Economy&source=fresco-home&unqvaldesktop=1000049930202"
driver.get(url)
time.sleep(2)
for i in range(200):
    driver.execute_script("window.scrollBy(0,1000)")
```

```python
Airline=[]
source=[]
destination=[]
Arival=[]
Departure=[]
no_of_stops=[]
prices=[]

name=driver.find_elements_by_xpath('//div[@class="fs-13 airline-name no-pad col-8"]/span')
for i in name:
    Airline.append(i.text)

try:
    so=driver.find_elements_by_xpath('//div[@class="i-b col-4 no-wrap text-right dtime col-3"]/p')
    for i in so:
        source.append(i.text)
except NoSuchElementException:
    source.append(None)

try:
    des=driver.find_elements_by_xpath('//div[@class="i-b pdd-0 text-left atime col-5"]/p[2]')
    for i in des:
        destination.append(i.text)
except NoSuchElementException:
    destination.append(None)
    
try:
    ari=driver.find_elements_by_xpath('//div[@class="i-b pr"]')
    for i in ari:
        Arival.append(i.text)
except NoSuchElementException:
    Arival.append(None)

try:
    dep=driver.find_elements_by_xpath('//div[@class="i-b pdd-0 text-left atime col-5"]//p[1]')
    for i in dep:
        Departure.append(i.text.split('\n')[0])
except NoSuchElementException:
    Departure.append(None)
        
st=driver.find_elements_by_xpath('//div[@class=" font-lightgrey fs-10 tipsy i-b fs-10"]')
for i in st:
    if i.text[0]!='N':
        no_of_stops.append(int(i.text[0]))
    else:
        no_of_stops.append(int('0'))

price=driver.find_elements_by_xpath('//div[@class="i-b tipsy fare-summary-tooltip fs-18"]')
for i in price:
    prices.append(i.text.replace('₹',''))

print(len(Airline))
print(len(source))
print(len(destination))
print(len(Arival))
print(len(Departure))
print(len(no_of_stops))
print(len(prices))
```

```python
df1= pd.DataFrame({"Airlines_Name":Airline,
                 "Date Of Journey":"27-Apr-2022",
                 "Source":source,
                 "Destination":destination,
                 "Arival":Arival,
                 "Departure":Departure,
                 "No. Of Stoppage":no_of_stops,
                 "Price":prices})

df1
```
